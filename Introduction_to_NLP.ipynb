## **Funtion to remove punctuation**

```python
def remove_punctuation(text):
    '''a function for removing punctuation'''
    import string


    # replacing the punctuations with no space,.
    # which in effect deletes the punctuation marks
    translator = str.maketrans('', '', string.punctuation)
    # return the text stripped of punctuation marks
    return text.translate(translator)
```

```python
remove_punctuation("shubhankit, is old")
```

## **Apply the function to each examples**

```python
data['text'] = data['text'].apply(remove_punctuation)
data.head(10)
```

## **Removing stopwords**

**Extract the stop words**

```python
import nltk
nltk.download('stopwords')
```

```python
# extracting the stopwords from nltk library
sw = stopwords.words('english')
# displaying the stopwords
np.array(sw)
```

```python
print("Number of stopwords: ", len(sw))
```

## **Function to remove stopwords**

```python
def stopwords(text):
    '''a function for removing the stopword'''
    
    # removing the stop words and lowercasing the selected words
    
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    # joining the list of words with space separator
    return " ".join(text)
```

```python
data['text'] = data['text'].apply(stopwords)
data.head(10)
```

##  **Stemming operations**

### Stemming operation bundles together words of same root. e.g. stem operation bundles "response" and "respond" into a common "respon"

**A funtion to carry out stemming operation**

```python
# create an object of stemming function
stemmer = SnowballStemmer("english")

def stemming(text):
    '''a function which stems each word in the given text'''
    text = [stemmer.stem(word) for word in text.split()]
    return " ".join(text)
```

**Apply the function to each examples**

```python
data['text'] = data['text'].apply(stemming)
data.head(10)
```

# **TF-IDF Extraction**

### tf-idf weight is product of two terms: the first term is the normalized Term Frequency (TF), aka. the number of times a word appears in a document, divided by the total number of words in that document; the second term is the Inverse Document Frequency (IDF), computed as the logarithm of the number of the documents in the corpus divided by the number of documents where the specific term appears.

### TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document).

### IDF(t) = $log_e(\frac{Total\  number \ of \ documents}{ Number \  of \ documents \  with \  term \ t \ in \ it })$

The TF-IDF vectorizer is a tool used in natural language processing to convert text data into numerical form. Here's a simple explanation of how it works:

### TF-IDF: Term Frequency-Inverse Document Frequency

1.  **Term Frequency (TF):**
    
    *   This measures how frequently a term (word) appears in a document.
        
    *   If a word appears more times in a document, its TF value is higher.
        
    *   The formula for TF is:TF(t,d)=Number of times term t appears in document dTotal number of terms in document d\\text{TF}(t, d) = \\frac{\\text{Number of times term } t \\text{ appears in document } d}{\\text{Total number of terms in document } d}TF(t,d)=Total number of terms in document dNumber of times term t appears in document d​
        
2.  **Inverse Document Frequency (IDF):**
    
    *   This measures how important a term is across all documents in the dataset.
        
    *   If a term appears in many documents, its IDF value is lower because it's considered less unique.
        
    *   The formula for IDF is:IDF(t)=log⁡(Total number of documentsNumber of documents with term t)\\text{IDF}(t) = \\log \\left( \\frac{\\text{Total number of documents}}{\\text{Number of documents with term } t} \\right)IDF(t)=log(Number of documents with term tTotal number of documents​)
        
3.  **Combining TF and IDF:**
    
    *   The TF-IDF score for a term in a document is calculated by multiplying its TF value with its IDF value.
        
    *   This score helps identify terms that are significant in a document but not too common across all documents.
        
    *   The formula for TF-IDF is:TF-IDF(t,d)=TF(t,d)×IDF(t)\\text{TF-IDF}(t, d) = \\text{TF}(t, d) \\times \\text{IDF}(t)TF-IDF(t,d)=TF(t,d)×IDF(t)
        

### Why Use TF-IDF?

*   **Identify Important Words:** It highlights words that are important in a specific document but not common in other documents.
    
*   **Improve Search Results:** Helps in finding documents relevant to a search query by focusing on significant words.
    
*   **Reduce Noise:** Filters out common words (like "the," "is," etc.) that don't carry much meaning.
    

### Example

Imagine you have three documents:

*   Document 1: "The cat sat on the mat."
    
*   Document 2: "The dog sat on the mat."
    
*   Document 3: "The cat chased the dog."
    


So, the word "cat" has a higher TF-IDF score in Document 3 than in Document 1, indicating that it is more significant in Document 3. This is how TF-IDF helps in understanding the importance of words within documents.
